# Popularity Based Recommendetion System 

```python
ratings_with_ISBN = ratings.merge(books,on="ISBN")
```

```python
ratings_with_ISBN.head()
```

```python
num_rating = ratings_with_ISBN.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating = num_rating.rename(columns = {"Book-Rating":"num_rating"})
num_rating.head()
```

```python
avg_rating = ratings_with_ISBN.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating =avg_rating.rename(columns = {"Book-Rating":"avg_rating"})
avg_rating.head()
```

```python
popularity_df = num_rating.merge(avg_rating,on="Book-Title")
```

```python
popularity_df.shape
```

```python
popularity_df = popularity_df[popularity_df["num_rating"]>=250].sort_values("avg_rating",ascending = False).head(50)
```

```python
popularity_df
```

```python
popularity_df = popularity_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")
```

```python
popularity_df = popularity_df.drop(columns = ["ISBN","Publisher","Year-Of-Publication","Image-URL-S","Image-URL-L"],axis = 1)
```

```python
popularity_df
```

# Collaborative Filtering Based Recommender System

```python
x = ratings_with_ISBN.groupby("User-ID").count()["Book-Rating"]>200
top_users = x[x].index
```

```python
filtered_rating = ratings_with_ISBN[ratings_with_ISBN["User-ID"].isin(top_users)]
```

```python
filtered_rating
```

```python
y=  filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
y
```

```python
famous_books = y[y].index
famous_books
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse = True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend("Message in a Bottle")
```

```python
recommend("The Notebook")
```

```python
recommend("1984")
```

```python
import joblib 
joblib.dump(similarity_score,"similarity_score.joblib")
joblib.dump(popularity_df,"popular.joblib")
joblib.dump(books,"Books.joblib")
joblib.dump(pt,"pt.joblib")
```
